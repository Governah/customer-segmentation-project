```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/sample.csv')
```

```python
df.head()
```

```python
from datetime import datetime

current_year = datetime.now().year

df["acct_open_date"] = pd.to_datetime(
    df["acct_open_date"],
    format="%m/%Y"
)
```

```python
df["account_age"] = (
    current_year - df["acct_open_date"].dt.year
)
```

```python
df[["acct_open_date", "account_age"]].head()
```

```python
df["pin_age"] = (
    current_year - df["year_pin_last_changed"]
)
```

```python
df[["year_pin_last_changed", "pin_age"]].head()
```

```python
df["expires_date"] = pd.to_datetime(
    df["expires_date"]
)
```

```python
df["years_to_expiry"] = (
    df["expires_date"].dt.year - current_year
)
```

```python
df[["expires_date", "years_to_expiry"]].head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["card_brand_encoded"] = le.fit_transform(
    df["card_brand"]
)
```

```python
df[["card_brand", "card_brand_encoded"]].tail()
```

```python
df[[
    "account_age",
    "pin_age",
    "years_to_expiry",
    "card_brand_encoded"
]].head()
```

```python
df["has_chip"] = df["has_chip"].map({
    "YES": 1,
    "NO": 0
})
```

```python
df["has_chip"].head()
```

```python
df["has_chip"].dtype
```

```python
df["card_type"].unique()
```

```python
df["card_type"] = df["card_type"].map({
    "Debit": 0,
    "Credit": 1
})
```

```python
df["card_type"].head()
```

```python
df["card_on_dark_web"].unique()
```

```python
df["card_on_dark_web"] = df["card_on_dark_web"].map({
    "No": 0,
    "Yes": 1
})
```

```python
df["card_brand"] = df["card_brand"].map({
    "America Express": 0,
    "Mastercard": 1,
    "Visa": 2
})
```

```python
df["card_on_dark_web"].head()
```

```python
df['credit_limit'] = df['credit_limit'].str.replace('$', '').astype(float)
```

```python
df['card_type'] = df['card_type'].fillna(1.0)
```

```python
df = df.drop(columns=['id', 'client_id', 'card_number', 'cvv', 'expires_date'])
```

```python
df['card_brand'] = df['card_brand'].fillna(df['card_brand'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['years_to_expiry'])
```
