```python
import pandas as pd

df = pd.read_csv("../data/sample.csv")
```

```python
df = df.drop(
    columns=[
        "id",
        "client_id",
        "card_number",
        "cvv"
    ]
)
```

```python
df["credit_limit"] = (
    df["credit_limit"]
    .replace(r"[$,]", "", regex=True)
    .astype(float)
)
```

```python
df["has_chip"] = df["has_chip"].map({
    "YES":1,
    "NO":0
})
```

```python
df["card_on_dark_web"] = df["card_on_dark_web"].map({
    "Yes":1,
    "No":0
})
```

```python
df["card_type"] = df["card_type"].map({
    "Debit":0,
    "Credit":1
})
```

```python
import datetime as dt

current_year = dt.datetime.now().year

df["account_year"] = pd.to_datetime(
    df["acct_open_date"],
    format="%m/%Y"
).dt.year

df["account_age"] = current_year - df["account_year"]
```

```python
features = [
    "credit_limit",
    "num_cards_issued",
    "has_chip",
    "card_type",
    "card_on_dark_web",
    "account_age"
]
```
